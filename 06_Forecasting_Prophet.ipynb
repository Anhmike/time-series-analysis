# Time Series Forecasting wih Prophet

Prophet is a procedure for forecasting time series data. It is based on an additive model where non-linear trends are fit with yearly and weekly seasonality, plus holidays. It works best with daily periodicity data with at least one year of historical data. Prophet is robust to missing data, shifts in the trend, and large outliers. - <a href="https://facebookincubator.github.io/prophet/">Prophet</a>

<code>Prophet</code> assume an additive model and thus, the data is transformed using the log operator.

```python
from fbprophet import Prophet
```

```python
df = pd.read_csv('./data/passengers.csv', sep=';', header=0, parse_dates=True)

# log difference data
#df['ds'] = df['month']
#df['y'] = pd.DataFrame(np.log(df['n_passengers']).diff())
#df.dropna(inplace=True)

df['ds'] = df['month']
df['y'] = pd.DataFrame(np.log(df['n_passengers']))
df.pop('month')
df.pop('n_passengers')

df.head()
```

```python
ax = df.set_index('ds').plot();
ax.set_ylabel('Passengers');
ax.set_xlabel('Date');

plt.show()
```

```python
# train test split
df_train = df[:120]
df_test = df[120:]
```

```python
mdl = Prophet(interval_width=0.95)
```

```python
# fit the model on the training data
mdl.fit(df_train)
```

```python
# define future time frame
future = mdl.make_future_dataframe(periods=24, freq='MS')
future.tail()
```

```python
# generate the forecast
forecast = mdl.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

<code>mdl</code> is a <code>Prophet</code> object and and can be plotted with predefined settings

```python
mdl.plot(forecast);
plt.show()
```

```python
# plot time series components
mdl.plot_components(forecast);
plt.show()
```

```python
# retransform using e
y_hat = np.exp(forecast['yhat'][120:])
y_true = np.exp(df_test['y'])

# compute the mean square error
mse = ((y_hat - y_true) ** 2).mean()
print('Prediction quality: {:.2f} MSE ({:.2f} RMSE)'.format(mse, math.sqrt(mse)))
```
